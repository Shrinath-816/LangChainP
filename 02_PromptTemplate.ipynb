# Prompt Templates

Definition - A Prompt Template is a reusable prompt structure that allows dynamic insertion of values into placeholders before sending the prompt to an LLM.

### Why Use Prompt Templates?
* Avoid hardcoded prompts
* Reusable and maintainable
* Dynamic user input handling
* Cleaner code
### Types of Prompts in LangChain
#### 1. String Prompt Template
Used for simple text-based prompts.
```python
template = "Explain {topic} in simple terms."
```
#### 2. Chat Prompt Template
Used for chat-based models with system, human, and AI messages.
```python
ChatPromptTemplate
```
### Key Components
* **Template** → Static prompt text
* **Input Variables** → Dynamic values
* **format()** → Replaces placeholders with actual values
### Workflow
Template + Variables → Formatted Prompt → LLM → Response
### Example
Template:
```python
"Explain {topic} in simple terms."
```
Input:
```python
topic = "Artificial Intelligence"
```
Final Prompt:
```text
Explain Artificial Intelligence in simple terms.
```
### Important Methods
```python
prompt.format()
```
Creates the final prompt by replacing placeholders.
### Interview One-Liner

PromptTemplate is a LangChain component that helps create dynamic and reusable prompts by injecting runtime values into predefined placeholders.

### 1 : String PromptTemplate

````python
from langchain_core.prompts import PromptTemplate
prompt_template = PromptTemplate.from_template("What is the capital of {country}?")
prompt = prompt_template.invoke({"country": "India"})
````

````python
from langchain_google_genai import ChatGoogleGenerativeAI
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0)
response = llm.invoke(prompt)
print(response)
````

###  Chat PrompTemplate
ChatPromptTemplate is a LangChain prompt template used to create structured conversations by defining different message roles such as system, human (user), and AI, before sending them to a chat model.

````python
from langchain_core.prompts import ChatPromptTemplate
chat_template = ChatPromptTemplate([("system", "You are a helpful assistant."),("user", "What is the capital of {country}?")])

prompt = chat_template.invoke({"country": "India"})
response = llm.invoke(prompt)
print(response.content)
````

### MessagesPlaceholder

MessagesPlaceholder is used inside a ChatPromptTemplate to dynamically insert a list of messages (typically chat history) at runtime.

````python
from langchain_core.prompts import ChatPromptTemplate , MessagesPlaceholder
from langchain_core.messages import HumanMessage

prompt_template = ChatPromptTemplate([
    ("system", "You are a helpful assistant."),
    MessagesPlaceholder("msg")
])

prompt = prompt_template.invoke({"msg": [HumanMessage(content="What is the capital of India?")]})
llm.invoke(prompt).content
````

````python
from langchain_core.prompts import ChatPromptTemplate , MessagesPlaceholder
from langchain_core.messages import HumanMessage, AIMessage

prompt_template = ChatPromptTemplate([
    ("system", "You are a helpful assistant."),
    MessagesPlaceholder(variable_name="msg")
])

# Define multiple messages
messages = [
    HumanMessage(content="What is the capital of India?"),
    AIMessage(content="The capital of India is New Delhi."),
    HumanMessage(content="What is the population of New Delhi?")
]

# Invoke the prompt template with the list of messages
output = prompt_template.invoke({"msg": messages})
print(output)
````

````python
llm.invoke(output).content
````
