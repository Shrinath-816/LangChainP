### Text Spliter:
A text spliter in Langchain is a utility that breaks large document or text into smaller manageable chunks while preserving context.These chunks are easier for LLM to process because they typically have token limits.

Spliting makes RAG , semantic search and summarization more efficient.

### Types of Text Spliters:
### a) Character TextSpliter : 
Splits text by character count.Cons - can break sentences mid way losing some context.

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
# RecursiveCharacterTextSplitter - Recursively tries to split text at natural boundaries (e.g. paragraphs, sentences, words) 


text = "Lancgchain makes it easy to build LLM-powerd app. It has text splitters to handle large documents."

ch_splitter = RecursiveCharacterTextSplitter(
    separators=[""],
    chunk_size=20,
    chunk_overlap=3
)

chunks = ch_splitter.split_text(text)

print(len(chunks))
print(type(chunks))
print(chunks[1])
```

```python
for i in chunks:
    print(i)
```

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

path = "D:\\LangChainP\\pdf_files\\Training and Learning Policy.pdf"
pdf_loader = PyPDFLoader(path)
pdf_docs = pdf_loader.load()

rec_spliter = RecursiveCharacterTextSplitter(
  separators=["\n\n", "\n", ".", " ",""], chunk_size = 500, chunk_overlap= 30
)

pdf_chunk = rec_spliter.split_documents(pdf_docs)

print(len(pdf_chunk))
print(pdf_chunk[1].page_content)
```

```python
for i in range(len(pdf_chunk)):
    print(pdf_chunk[i].page_content)
```

### MarkdownHeader Text Spliter

```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import MarkdownHeaderTextSplitter

path = "text.md"

md_laoder = TextLoader(path)
md_docs = md_laoder.load()

md_text = md_docs[0].page_content

md_spliter= MarkdownHeaderTextSplitter(
  headers_to_split_on=[("#","Title"), ("##", "Section"), ("###","Subsection")]
)

md_chunks = md_spliter.split_text(md_text)

print(len(md_chunks))
print(md_chunks[0].page_content)
```

```python
print(md_chunks[3].page_content)
```

```python
for i in md_chunks:
  print(i.page_content)
```
