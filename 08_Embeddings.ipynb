### Embedding : 

An embedding is a numerical represenation of text(sentence , paragraph , documents) as a vector in a high dimensional space(eg. 384 - 3072+ dimensions).

The idea : similar meaning ---> similar vector

### Key Uses : 

1) Semantic search : find relevant text even if exact words dont match.
2) RAG:
3) Clustering / Classification : group similar texts together.
4) Duplication / Paraphrase Detection : Identify near duplicate content.
5) Recommendation : Suggest items based on the semantic similarity.


```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from dotenv import load_dotenv

load_dotenv()

embed = GoogleGenerativeAIEmbeddings(model="gemini-embedding-001")

#embed.embed_query("I live in India")
```

```python
vector = embed.embed_query("I live in India")
print(len(vector))
print(vector[:10])
```

### Embed PDF File

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from dotenv import load_dotenv
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
load_dotenv()

path = "D:\\LangChainP\\pdf_files\\Training and Learning Policy.pdf"
loader = PyPDFLoader(path)
docs = loader.load()

splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
chunks = splitter.split_documents(docs)

text_content = [chunk.page_content for chunk in chunks]
embed = GoogleGenerativeAIEmbeddings(model="gemini-embedding-001")
#embed.embed_documents(text_content)

```

```python
vector = embed.embed_documents(text_content)
print("chunk size:", len(vector))
print("dimensions:", len(vector[0]))
```
