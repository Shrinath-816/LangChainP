### Document Loaders
A document loader in Langchain is a utility that helps you ingest data from external source(files , apis ,database) and converts it into a Langchain Document Object.


Document = Page content + metadata

Category :


a) File based loaders : for local file or cloud stored files. PyPDFLoader : whichs loads PDFs page by page, TextLoader : loads plain text, CSVLoader : loads tabular CSV file etc.

b)Web based Loaders : For scraping or API-based ingestion. WebBasedLoader : loads web pages using requests, SitemapLoader : loads websites using sitemap.xml

c) Database/API loaders : for pulling structured data from database & APIs.  DataFrameLoader : converts Pandas DataFrame to documents,  SQLLoader : loads from SQL database,  FAISS/Weaviate Loaders : Load from vector database,  APIResponseLoader : Parse API response into documents.

d) Structured Data Loaders : for files with schemas or complex structure.  StructureFiledLoader : advanced CSV/JSON/Excel with schema , ExcelLoader : Specifically for .xlsx files , MarkDownLoader : Extracts text & headers from .md files

### Text Loader

TextLoader loads plain .txt files into Langchain documents.

```python
from langchain_community.document_loaders import TextLoader

path = "requirements.txt"

loader = TextLoader(path)

docs = loader.load()

# print(len(docs))
# print(type(docs))
print(docs[0].page_content)
```

### PyPDFLoader
PyPDFLoader is a document loader in Langchain that extracts text from PDF files page by page.cUnlike simple text extraction, PyPDFLoader ensures that each page becomes a separate document making downstream chunking and embedding easier.

```python
from langchain_community.document_loaders import PyPDFLoader

path = "data/Training and Learning Policy.pdf"

pdf_loader = PyPDFLoader(path)

docs = pdf_loader.load()

print(len(docs))
print(docs[0].page_content)
```

### PyPDFDirectoryLoader 
We use PyPDFDirectoryLoader in LangChain to automatically ingest and parse all PDF files within a specific local directory. systems when your knowledge base consists of multiple PDF documents rather than just one.

### Why Use PyPDFDirectoryLoader?
### Batch Processing: 
Instead of loading PDFs one by one, it processes an entire folder with a single command, saving time and code.
### Automatic Metadata: 
It automatically captures file details (like the source file name and page number) and adds them to the parsed document chunks.
### Lightweight & Fast: 
It uses the native Python pypdf library, meaning it doesn’t require heavy external dependencies or API keys to extract raw text.
### Recursive Search:
 You can configure it to search through subfolders to keep large, nested document repositories organized.

```python
from langchain_community.document_loaders import PyPDFDirectoryLoader

multipdf_laoder = PyPDFDirectoryLoader(
  path = "D:\LangChainP\pdf_files", glob = "**/*.pdf"
)

multipdf_docs = multipdf_laoder.load()

print(len(multipdf_docs))
print(multipdf_docs[0].page_content)
```

### CSVLoader :
loads structured tabular data from .csv files into Langchain.Each row of the CSV becomes a separate Document with columns stored as metadata.

```python
from langchain_community.document_loaders import CSVLoader

path = "data\\ecomm.csv"

csv_loader = CSVLoader(path)

csv_docs= csv_loader.load()

print(len(csv_docs))
print(csv_docs[0].page_content)
```

### JSON Loader

```python
from langchain_community.document_loaders import JSONLoader

path = "data\sample.json"

json_laoder = JSONLoader(path, jq_schema=".", text_content=False)

json_docs = json_laoder.load()

print(len(json_docs))
print(json_docs[0].page_content)
```

### Structured File Loader:
StructuredFileLoader is a generic loader for structured files like CSV , Excel or JSON where you want to apply a schema or a parser.Its more advanced than CSVLoader or JSONLoader because it gives you flexibility in handling different formats and allows schema based validation before creating documents.

```python
from langchain_community.document_loaders import UnstructuredFileLoader, JSONLoader
from pathlib import Path

files = Path("data").glob("*")

all_docs = []

for file in files:
  if file.suffix == ".json":
    unstructured_loader = JSONLoader(file, jq_schema=".", text_content=False)
  else:
    unstructured_loader = UnstructuredFileLoader(file)
  all_docs.extend(unstructured_loader.load())

print(len(all_docs))
print(all_docs[0].page_content)
```
